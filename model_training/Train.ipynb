```python
import torch
from ultralytics import YOLO

# Set the device to MPS if available
device = 'mps' if torch.backends.mps.is_available() else 'cpu'

model = YOLO('yolo11n-pose.pt')
config_file = 'config.yaml'

# Documentation on how to training the model
# https://docs.ultralytics.com/modes/train/

model.train(
    device=device,
    data=config_file,
    epochs=50,           
    imgsz=640,            
    batch=8,              
    workers=8,
    name='example_results',
    exist_ok=False,
    cache=True,
    # Loss weights
    box=7.5,
    cls=0.5,
    dfl=1.5,              
    # Augmentations:
    hsv_h=0.015,          
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=10.0,         
    scale=0.1,               
    shear=0.0,
    flipud=0.5,
    fliplr=0.5,
    mixup=0.0,             
    copy_paste=0.0
)









```
